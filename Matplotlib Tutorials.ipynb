### Lesson 1 - Introduction

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='red', linewidth=5, linestyle='dotted')
```

### Lesson 2 -  Format String in plot function

```python
plt.plot(x, y, color = 'green' , marker= '+', markersize=15, alpha=0.8)  #alpha is for opacity
```

### Lesson 3 - Axes Lables, Legend, Grid

```python
days= [1,2,3,4,5,6,7]
max_t= [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Chart')
plt.plot(days, max_t, color='green', marker='D', markersize=6, label='Max')
plt.plot(days, min_t, color='red', marker='D', markersize=6,label='Min')
plt.plot(days, avg_t, color='orange', marker='D', markersize=6,label='Avg')
plt.legend(loc="best", shadow=True, fontsize="small")

#Grid
plt.grid()
```

### Lesson 4 - Bar Chart

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue= [90, 136, 89, 27]
profit =[40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel('revenue(bin)')
plt.xlabel('Company')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, color='green', width=0.4,label='Revenue')
plt.bar(xpos+0.2, profit, color='red',width=0.4,label='Profit')
plt.legend()
```

### Lesson 5- Pie Chart

```python
exp_vals =[1400,600,300,410,250]
exp_labels =["Home Rent","Food", "Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels= exp_labels, radius =1.5, autopct='%0.2f%%', explode=[0,0.3,0,0,0])
```
